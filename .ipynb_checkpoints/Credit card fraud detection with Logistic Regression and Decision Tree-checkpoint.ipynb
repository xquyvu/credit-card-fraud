# Part 1. Analysis domain, questions, plan:

## 1.1. Introduction:

In the UK only, fraud losses on card amounted to £618.0 million in 2016, a nine-percent increase from £567.5 million in 2015. This figure has been increasing for 5 years, showing itself being an alarming issue. (Financial Fraud Action UK, 2017). In September 2017, 702.762 transactions were made [1] - a huge volume that poses enormous losing risk if verification is postponed.

However, "an institution may have state-of-the-art security and fraud detection technologies and procedures to protect its deposit lines of business, but not the same for small business banking or third-party investments delivered by an alliance partner,” [2].

As such, this project aims to develop a framework that can detect fraud in credit card transactions with high accuracy and is computationally efficient and thus can be widely implemented.

## 1.2. The dataset: 

The dataset was retried from kaggle (https://www.kaggle.com/ntnu-testimon/paysim1/data) and was generated using a simulator that uses aggregated data from a private dataset to generate a synthetic dataset that resembles the normal operation of transactions. The original dataset were provided by a multinational company, who is the provider of the mobile financial service which is currently running in more than 14 countries all around the world.

The generated dataset contains more than 26 million entries. However, only the first 1,030,564 entries representing transactions made within 2 days were analysed due to limited computing power. If promising result is obtained, further analyses will be performed on the whole dataset.

The dataset has 10 features containing information on the time occurred, type of transactions (payment, transfer, cash in, ...), amount, balance of customer and recipient before and after the transaction, the class of the transactions with representing fraud and 0 if otherwise, and whether the transaction is flagged as fraud by a system or not.

Since only 890 transactions were classified as fraud amongst the 1,030,564, the dataset is highly skewed.

There is no missing values

## 1.3. Analysis Strategy and Plans:

**1.3.1 Problem formulation**

Since the aim of this project is to predict whether a transaction is fraudulent or not, the task is defined as classification.

**1.3.2. Data pre-processing**

- _**Data imbalance**_: As analysed, the dataset is highly skewed. As such, the rare class, i.e fraud, are oversampled with random Gaussian noises being added to address this problem.

- _**Missing values**_: There is possibly missing values encoded as 0. Imputation will be considered.

- _**Categorical data**_: The dataset also contains information on (i) the type of transaction, (iv) the time when the transaction occurred(iii) the name of customers and recipients, and (iv) classification made by the system. (i) and (ii) will be transformed into binary values with OneHotEncoder. We will assess the relevance of (iii) and (iv) and decide to discard or not.

- _**Data normalisation**_: Besides "step" specifying the time when the transaction took place, all other numerical features represents an amount of money. However, these values are not on the same scale. Data normalisation will be considered based on the selected model.

**1.3.3. Modelling**

To quickly produce an interpretable prototype that helps knowledge discovery, logistic regression and decision tree were selected.

**1.3.4. Evaluation**
<br>

Performance on the test set, AUC, confusion matrix will be selected as the main evaluating criteria.

**1.3.5. Analysis aim**
<br>
The goal of this project is to maximize the true positive rate (a fraudulent transaction being flagged as fraud) in exchange of some false negative rate where necessary. This is because in a business context, succesfully detect a fraudulent transaction can save money from being stolen, while misclassify a legit transaction only requires users to undergo a quick process to verify their transactions.

## 2.2. Data preprocessing

### 2.2.1 Feature engineering

#### 1. Discarding "isFlaggedFraud"

As analysed, "isFlaggedFraud" was considered irrelevant and will be discarded.

```python
data = data.drop('isFlaggedFraud', axis = 1)
```

#### 2. Examining the relevance of customer's and recipient's name

We will see if names occurred more than once in fraudulent transactions.

```python
from collections import Counter

# Create a list of fraudulent transactions
name_fraud = data[data.isFraud == 1]

# Create a list of repeated name of customers in fraudulent transactions
nameOrig_fraud = name_fraud.nameOrig
nameOrig_counter = Counter(nameOrig_fraud)
nameOrig_repeated = Counter(i for i in nameOrig_counter.elements() if nameOrig_counter[i] > 1)

# Create a list of repeated name of recipients in fraudulent transactions
nameDest_fraud = name_fraud.nameDest
nameDest_counter = Counter(nameDest_fraud)
nameDest_repeated = Counter(j for j in nameDest_counter.elements() if nameDest_counter[j] > 1)

print('Repeated customer names in fraudulent transactions:') 
print(nameOrig_repeated)
print('')
print('Repeated customer names in genuine transactions:') 
print(nameDest_repeated)
```

From the above, only recipient names occurred more than once with the highest value being 2. One can imply that a fraudulent behavior is to first transfer the money to an account before withdrawing it, the 2 occurences represent these 2 actions. As a result, we can consider such names to be insignificant and can be removed.

```python
# Removing the names of customers and recipients
data = data.drop(['nameOrig', 'nameDest'], axis = 1)
```

#### 3. Discarding transaction types other than "TRANSFER" and "CASH_OUT":

As discussed, only transfering and cash out transactions are considered. Transactions of other types are thus discarded.

```python
# Discarding transaction types other than "TRANSFER" and "CASH_OUT"
data = data.loc[(data.type == 'TRANSFER') | (data.type == 'CASH_OUT')]
data = data.reset_index(drop=True)
```

#### 4. Factorising "step" column:

The "step" column represents the time when the transaction took place, ranging from 1 representing the first hour in the first day to 72 representing the 24th hour of the third day. Taking this into account, "step" was paritioned into "hour" and "date", ranging from 0 to 23 and 1 to 3 respectively. This factorisation is only to facilitate exploratory data analysis. 

```python
# Adding 'hour' and 'date' columns, removing 'step' column
data.step = data.step -1
data['hour'] = data.step % 24
data['date'] = data.step // 24 +1
data = data.drop('step', axis = 1)
# data.head()
```

### 2.2.2. Data cleaning

**1. Destinator's balance equals to zero revisited**

We now investigate the issue of the destinator being zero after the transactions. This could possibly mean that the transaction is halted or due to missing values. We will not increase newbalanceDest by the transaction amount in non-fraudulent transactions, and maintain zero values in fraudulent transactions, as this will give away the transaction label. We will consider the necessary of imputation by examining the significance of this finding.

```python
fraud_zero = np.sum(np.logical_and(np.logical_and(data.isFraud == 1,data.newbalanceDest == 0),data.oldbalanceDest == 0))
clean_zero = np.sum(np.logical_and(np.logical_and(data.isFraud == 0,data.newbalanceDest == 0),data.oldbalanceDest == 0))

print('The number of fraudulent transactions with newbalanceDest equals to zero is ' + str(fraud_zero) +', which is ' +str(round(fraud_zero / sum(data.isFraud)* 100,2)) + ' %.')
print('The number of clean transactions with newbalanceDest equals to zero is '      + str(clean_zero) +', which is ' +str(round(fraud_zero / len(data) * 100,2)) + ' %.')
```

The results have clearly shown that destinator's balances being zero is a strong indicator of fraud. As a common approach, we will impute a value -1 in oldbalanceDest and newbalanceDest of observations where this is true.

```python
# Imputing
data.loc[(data.oldbalanceDest == 0) & (data.newbalanceDest == 0) & (data.amount != 0), ['oldbalanceDest', 'newbalanceDest']] = - 1
```

**2. Encoding transaction type**

The 2 remaining types of transactions will be encoded as 0 and 1 respectively so models taking only numerical values as input can be applied to this dataset. 

At the end of the data cleaning stage, the update dataset is as follow:

```python
# encoding transaction type
data.loc[(data.type == 'CASH_OUT'), ['type']] = 1
data.loc[(data.type == 'TRANSFER'), ['type']] = 0
data.head()
```

## 2.4. Modelling

### 2.4.1. Model selection

Logistic regression and decision trees were selected for the following reasons:

- The models do not rely on many assumptions which are not true for this dataset, such as normally distributed data, linear relationship, homogenity of variance. [4]
- Computational cost is not expensive comparing to other classifiers. [5]
- Ease of interpretation

### 2.4.2. Data preparation
**1. Encoding "hour" and "day" features**

To prepare for the modelling step, these two features were encoded as 0 and 1, because they represented different points in a timeline instead of time duration. This is more problematic for logistic regression, as the algorithm returns the parameters for the hypothesis function $z_{\theta}(x) = \theta^T x$. If $\theta_{hour}$ or $\theta_{day}$ is larger than zero, the probability for a transaction being fraudulent will increase as time progresses and vice versa. This is an incorrect interpretation. 

However, creating a new dummy variable for each value in "hour" would add 23 dimensions to the dataset, each with over 465 thousands rows. This would result in expensive computation, as well as inefficient computation as most of the values in these dummy variables will be zero. Sparse matrix calculation approach can be applied, but for simplicity, a day was patitioned into 8 3-hour blocks.

After converting "hour" and "date" into dummy variables, one of each generated set of dummy variables are removed to avoid the dummy variable trap.

```python
# Creating 8 blocks of hour, dropping "hour".
data['h_block'] = np.ceil((data.hour+1) / 3)
data = data.drop('hour', axis = 1)

# Encoding h_block and date
date_enc = pd.get_dummies(data.date)
date_enc = date_enc.iloc[:,0:2]
date_enc.columns = ['d1', 'd2']

hour_enc = pd.get_dummies(data.h_block)
hour_enc = hour_enc.iloc[:,0:-1]
hour_enc.columns = ['h1', 'h2', 'h3', 'h4', 'h5','h6','h7']

# dropping "date" and "h_block"
data = data.drop(['h_block','date'], axis = 1)

# Joining them together
data = data.join([date_enc,hour_enc])
data.head()
```

**2. Normalising data**

In general, data normalisation speeds up convergence of algorithms supported by a certain kind of optimisation techniques and prevents them from being held in a local optimum. 

As logistic regression usually needs a regularisation terms to be protected against overfitting, the dataset was rescaled to have zero mean and unit standard deviation. Normalisation was be applied to binary categorical features, such as "type", "date", and "hour", as it would not benefit the interpretability of logistic regression.

**3. Splitting the data into training set and test set**

80% of the dataset will be used for training and the rest will be used for testing. The partitioning will be applied after the data is shuffled to minimise bias.

```python
# Specifying predictors and output 
X = standard_scaler.fit_transform(data.iloc[:,1:6])
data.iloc[:,1:6] = X
X = data.drop('isFraud',axis =1)
Y = data.isFraud

from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

**4. Oversampling the dataset**

As discussed, this dataset is highly imbalance. To address this issue, we will apply an oversampling technique, which is SMOTE.

```python
from imblearn.over_sampling import SMOTE

oversampler = SMOTE(random_state=0)
os_Xtrain,os_Ytrain = oversampler.fit_sample(Xtrain,Ytrain)
```

### 2.4.3. Model training

Both models were trained using the oversampled data, and the optimal parameters will be chosen based on the AUC score.

**1. Logistic regression**

Logistic regression was accompanied with a L2 regularisation term with different values of regularising parameters being: 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000. In scikit-learn, this indicates the inverse of the regularisation strength with smaller values correspond to stronger regularisation.

**2. Decision tree**

As a rule of thumbs, the minimum number of samples required to split an internal node is 1% of the dataset, which was 4544. For interpretation purpose, different values for the minimum number of leaves and maximum depth were [1:4] and [4:7] respectively. Splitting criterion was "best split at each node".

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, precision_recall_curve,auc,roc_auc_score,roc_curve,recall_score,classification_report
import itertools

cparams = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]
lr_results = pd.DataFrame(columns = cparams)

for v in cparams:
    lr = LogisticRegression(fit_intercept = False, C = v, penalty = 'l2')
    lr.fit(os_Xtrain, os_Ytrain)
    Ypred = lr.predict(Xtest) 
    lr_results.loc[0,v] = roc_auc_score(Ytest, Ypred)
```

Both models produced good results, and the AUC score being steep then flattened out suggested that variation of decision threshold is not necessary, as the trade-off would be significant.

Logistic regression produced decent results, with the highest AUC score being 0.948. Also, higher regularising parameters correspond to better results, meaning that the model does not heavily suffered from overfitting.

```python
fpr, tpr, thresholds = roc_curve(Ytest, Ypred)
roc_auc = auc(fpr,tpr)

# Plot ROC
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b',label='AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.0])
plt.ylim([-0.1,1.01])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

Normalised confusion matrix would be more helpful in assessing imbalance data. This simple model misclassified only 6% of the fraudulent transactions, which is tolerable.

```python
# Plotting confusion matrix, taken from scikit learn 
# http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Oranges):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=0)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
cnf_matrix = confusion_matrix(Ytest, Ypred)
np.set_printoptions(precision=2)

class_names = ['Genuine','Fraudulent']

# Plot non-normalized confusion matrix
figure9 = plt.figure(figsize = (15,4))
ax = figure9.add_subplot(121)
ax.grid()
plot_confusion_matrix(cnf_matrix, classes=class_names, title='Confusion matrix, without normalisation')

# Plot normalized confusion matrix
ax = figure9.add_subplot(122)
ax.grid()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True, title='Confusion matrix, with normalisation')

plt.show()
```

**2. Decision tree**

The optimal decision tree has 1 leaf minimum and 7 maximum depth, which produces an AUC score of 0.97.

```python
from sklearn import tree
dt_param1 = [1,2,3,4]
dt_param2 = [4,5,6,7]
optimal_dt_auc = 0
optimal_min_samples_leaf = 0
optimal_max_depth = 0
for i in dt_param1:
    for j in dt_param2:
        clf = tree.DecisionTreeClassifier(splitter = 'best', min_samples_split = 4544, min_samples_leaf = i, max_depth = j, random_state = 0)
        clf = clf.fit(os_Xtrain, os_Ytrain)
        dt_Ypred = clf.predict(Xtest)
        fpr2, tpr2, thresholds = roc_curve(Ytest, dt_Ypred)
        roc_auc = auc(fpr2,tpr2)
        if roc_auc > optimal_dt_auc:
            optimal_dt_auc = roc_auc
            optimal_min_samples_leaf = i
            optimal_max_depth = j
            
print('The optimal decision tree has ' + str(optimal_min_samples_leaf) + ' leaves minimum and ' + str(j) + ' maximum depth with an AUC score of ' + str(roc_auc))
```

# Part 3. Conclusions and future work

Financial data, especially crime-related are extremly scarce due to confidentiality, which somehow limits research feasibility. 

This work has discovered interesting patterns that help identifying frauds. Firstly, the analysis has highlighted the period of time when an occurring transaction is likely to be fraudulent, which is night time. Secondly, a combination of information on balance of originator and destinator served as a strong indicator for fraud. Although this combination could be hard to interpret by human, it is well captured by machine learning models, in this case being decision tree. Thirdly, a common behavior in fraudulent transaction is to first transfer the money from the victim's account then cash out. It is worth noting that the analysis was performed on a dataset with many missing values, thus a more complete one will indeed provide more insights.

The initial objective of building a interpretable, computationally efficient and reliable has been met. The method utilised in this work facilitates interpretation, thus hastened knowledge discovery, and at the same time consume minimal computational resource. Also, the prediction output was fruitful with decision tree being the bestter performing model managed to detect 99% of fraudulent transactions. Therefore, the method implemented in this work is widely applicable. 

Nevertheless, this work has some limitations, thus future work can be done. In specific:

- Only the first 1,030,564 entries were considered in this dataset. Analysing the whole dataset may give additional useful insights that can boost the model's performance.


- In this analysis, customers' and recipients' names were removed from the original dataset. However, by analysing the transaction history of such names, interesting insights may be noticed.


- Since fraudulent transactions only occurred in transfering and cash out transactions, this analysis only considered these 2 types. Future work can be done by creating an anomaly detection system that is trained on all types of transactions, and if properly built and tested on additional data, it can flag abnormal transactions as potential fraud and draw reasonable attention to those transactions.


- The optimisation technique was trying different values for parameteres. Other sophisticated approaches such as gradient descent, Newton's method could be considered, but in exchange of computational cost.


- Logistic regression and decision tree produced fruitful results. Thus, one may gain marginal improvement with other state-of-the-art algorithms.
